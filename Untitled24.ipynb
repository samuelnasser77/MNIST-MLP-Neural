```python
# Libraries
import numpy as np
import matplotlib.pyplot as plt

from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Dropout, BatchNormalization
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.optimizers import Adam

# Load Dataset
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()

print("Train Shape:", train_images.shape)
print("Test Shape:", test_images.shape)

train_images = train_images / 255.0
test_images = test_images / 255.0

train_labels = to_categorical(train_labels, 10)
test_labels = to_categorical(test_labels, 10)
# Model Builder Function
def build_model(neurons=128, dropout_rate=0.3, lr=0.001):
    model = Sequential()

    model.add(Flatten(input_shape=(28, 28)))

    model.add(Dense(neurons, activation='relu'))
    model.add(BatchNormalization())
    model.add(Dropout(dropout_rate))

    model.add(Dense(neurons // 2, activation='relu'))

    model.add(Dense(10, activation='softmax'))

    model.compile(
        optimizer=Adam(learning_rate=lr),
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )

    return model
# Experiment 1
print("\nTraining Model 1 (Baseline)...")

model1 = build_model(neurons=128, dropout_rate=0.3, lr=0.001)

history1 = model1.fit(
    train_images, train_labels,
    epochs=10,
    batch_size=32,
    validation_split=0.2,
    verbose=1
)

loss1, acc1 = model1.evaluate(test_images, test_labels)

print(f"\nModel 1 Accuracy: {acc1:.4f}")
print(f"Model 1 Loss: {loss1:.4f}")
# Experiment 2
print("\nTraining Model 2 (Improved)...")

model2 = build_model(neurons=256, dropout_rate=0.4, lr=0.0005)

history2 = model2.fit(
    train_images, train_labels,
    epochs=10,
    batch_size=32,
    validation_split=0.2,
    verbose=1
)

loss2, acc2 = model2.evaluate(test_images, test_labels)

print(f"\nModel 2 Accuracy: {acc2:.4f}")
print(f"Model 2 Loss: {loss2:.4f}")
# Visualization Function
def plot_history(history, title):

    # Accuracy
    plt.figure()
    plt.plot(history.history['accuracy'])
    plt.plot(history.history['val_accuracy'])
    plt.title(f'{title} Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend(['Train', 'Validation'])
    plt.show()

    # Loss
    plt.figure()
    plt.plot(history.history['loss'])
    plt.plot(history.history['val_loss'])
    plt.title(f'{title} Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend(['Train', 'Validation'])
    plt.show()


plot_history(history1, "Model 1")
plot_history(history2, "Model 2")
# Comparison
print("\n========== FINAL COMPARISON ==========")
print(f"Model 1 Accuracy: {acc1:.4f}")
print(f"Model 2 Accuracy: {acc2:.4f}")

if acc2 > acc1:
    print("Model 2 is better ✅")
else:
    print("Model 1 is better ✅")
# Prediction Example
import random

index = random.randint(0, len(test_images) - 1)
sample = test_images[index]

prediction = np.argmax(model2.predict(sample.reshape(1, 28, 28)))

plt.imshow(sample, cmap='gray')
plt.title(f"Predicted: {prediction}")
plt.axis('off')
plt.show()
# Save Best Model
best_model = model2 if acc2 > acc1 else model1

best_model.save("best_mnist_mlp_model.h5")

print("\nBest Model Saved Successfully")
```
